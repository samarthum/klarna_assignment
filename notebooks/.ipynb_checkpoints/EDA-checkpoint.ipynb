**Read Data**

```python
train = pd.read_csv('../data/train.csv')
```

**Data Exploration and checks**

**The target variable** <br>
The goal here is to build a model that can predict the probability of a loan default given a set of features. Before we get into more EDA and model building, we'll do a quick exploration of the target variable, or the dependent variable, 'default'.

```python
train['default'].value_counts().plot(kind='bar');
display(train['default'].value_counts(normalize=True))
```

As expected in such problems, the target variable is heavily imbalanced. The loan defaults occur only **1.43%** of the times. 
We can either treat this as an **imbalanced classification** problem, or we can treat it as an **anomaly detection** problem.

Before diving into further data exploration, let's quickly train a model and try to get an idea of feature importances from it. We can then do deep dives into individual features, and maybe engineer a few features as well.

For that, we will first have to take care of the missing values, as well as the categorical features.

**Correlation Heatmap** <br>
Let's take a look at how features are generally correlated amongst themselves, as well as with the target feature.

```python
plt.figure(figsize=(20, 15))
sns.heatmap(train.corr(), annot=True);
```

One of the things that stands out very clearly is that, none of the features are strongly correlated with the target variable 'default'. This is not good news, but we can see how to deal with this. 

The other thing that stands out is that, there are clusters of features that seem to be correlated to each other. This is probably due to the fact that they're all measuring a similar quantity across different time windows. A dataset of this sort could be a good candidate for dimensionality reduction because we could capture the information contained in these features using a smaller number of features.

**Missing Values**

```python
# Get sub-dataframe of just the missing columns

missing = train.isnull().mean()
cols_with_missing = missing[missing > 0].index

display(missing[missing > 0].sort_values(ascending=False))

missing = train[cols_with_missing]
```

```python
msno.heatmap(train);
```

We notice that a lot of the missing values are actually correlated. This indicates that the missing values are not Missing Completely At Random (MCAR). 
We can try to fill in these missing values in some meaningful way. Let's dig in deeper.

```python
msno.matrix(train[cols_with_missing]);
```

There could be some pattern of missing values wrt to the target variables as well. We can compare the above plot for the defaulters and non-defaulters. 

```python
defaults = train[train['default'] == 1]
non_defaults = train[train['default'] != 1]
```

```python
msno.matrix(defaults[cols_with_missing]);
```

```python
msno.matrix(non_defaults[cols_with_missing]);
```

There is definitely some difference of the distribution of null values between defaulters and non-defaulters. We can try to fill in the values in an effective manner by digging in a bit more. 

```python
# From problem description 

categorical_cols = ['default', 'account_status', 'account_worst_status_0_3m', 'account_worst_status_12_24m',
                    'account_worst_status_3_6m', 'account_worst_status_6_12m', 'merchant_category',
                    'merchant_group', 'name_in_email', 'status_last_archived_0_24m', 
                    'status_2nd_last_archived_0_24m', 'status_3rd_last_archived_0_24m',
                    'status_max_archived_0_6_months', 'status_max_archived_0_12_months',
                    'status_max_archived_0_24_months', 'worst_status_active_inv', 'has_paid']

numeric_cols = train.drop(categorical_cols, axis=1).columns.tolist()
```

```python
for col in cols_with_missing:
    if col in categorical_cols:
        plt.figure()
        plt.title(col)
        sns.histplot(data=train, x=col)
        plt.close();
        
    if col in numeric_cols:
        plt.figure()
        plt.title(col)
        sns.displot(data=train, x=col)
        plt.close();
```
